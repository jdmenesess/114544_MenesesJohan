```python
#decisiones y ciclos (ejercicio 7)
x = (int)(input('digite un numero: '))
x = abs(x)
a = []
while (x >= 10):
    res = x%10
    a.append(res)
    x = x//10
a.append(x)#convertimos el numero a lista
a.reverse()#lo re_organizamos
print(a)
b = []#creamos una lista vacia
for i in range (0,len(a)):#iteramos dentro del rango de la lista a 
    for j in range(i+1,len(a)):#hacemos una iteracion dentro de la iteracion
        #                       para poder comparar los elementos de la lista
        if a[i] == a[j]:#si el elemtno de la primera iteracion es igual al 
            #            elemento de la segunda iteracion
            b.append(a[i])#sumamos este elemento a la lista vacia definida con 
            #              anterioridad
for i in range (0,len(b)):#hacemos otra iteracion dentro de una iteracion
    for j in range(i+1,len(b)):#para poder eliminar los elementos repetidos
        #                       en la liusta b
        if b[i] == b[j]:#al igual que como en la anterior iteracion comparamos
            #           los elementos de la lista
            b[j] = 'repit'#si hay un numero que se repite vamos a cambiar el 
            #             elemento por una cadena de texto
c = []#creamo una lista vacia
for i in range(0,len(b)):#iteramos en el rango de la lista b
        if b[i] != 'repit':#y los elementos que sean diferentes de 'repit'
            c.append(b[i])#van a ser sumados a la nueva lista vacia
print('los digitos que se repiten son: ',c)#de esta manera encontramos los
    #                                       digitos repetidos en el numero
    #                                       ingresado al inicio
```

```python
#decisiones y ciclos (ejercicio 9)
import math#importamos la libreria math
#definimos las entradas para las cordenadas y el radio de los dos circulos
x1=(int)(input('(primer circulo)introduzca la coordenada en x: '))
y1=(int)(input('(primer circulo)introduzca la coordenada en y: '))
r1=(int)(input('(primer circulo)introduzca el radio: '))

x2=(int)(input('(segundo circulo)introduzca la coordenada en x: '))
y2=(int)(input('(segundo circulo)introduzca la coordenada en y: '))
r2=(int)(input('(segundo circulo)introduzca el radio: '))
#definimos las coordenadas (a,b) de un punto
a=(int)(input('coord en a: '))
b=(int)(input('coord en b: '))
#aqui definimos la distancia del punto (a,b) a las cordenadasde los 2 circulos
dist1=math.sqrt((x1-a)**2+(y1-b)**2)
dist2=math.sqrt((x2-a)**2+(y2-b)**2)

cont=0
if (dist1<=r1):# si la distancia es menor al radio del circulo 1
    print('(a,b) esta en el circulo 1')#significa que el punto esta dentro
    #                                   del circulo 1
    cont=cont+1#iniciamos un contador para este caso
else:#si no se cumple la condicion anterior el punto no esta en el circulo 1
    print('(a,b) no esta en el circulo 1')
if (dist2<=r2):#si la distancia es menor al radio del circulo 2 
    print('(a,b) esta en el circulo 2')#significa que el punto esta dentro del
    #                                   circulo 2
    cont=cont+1#iniciamos un contador para este caso
else:#si no se cumple la condicio anterior el punto no esta en el circulo 2 
    #ninguno de los dos circulos
    print('(a,b) no esta en el circulo 2')
if (cont==2):#si nuestro contador es igual a 2 significa que cumplio las dos 
    #         condiciones anteriores por lo tanto el punto se encuentra en los
    #         2 circulos
    print('(a,b) esta en los dos circulos')
if (cont==0):#si el contador es 0 significa que que no se cumplio las dos 
    #         condiciones anteriores por lo tanto el punto no esta en ninguno
    #         de los dos circulos
    print('(a,b) no esta en ninguno de los dos circulos')
```

```python
#cadenas (ejercicio 10)
cadena = (input('cadenas de texto: '))#definimos una entrada para una cadena 
#                                      de texto
len(cadena)
a = ord('a')#definimos el codigo UUNICODE para el caracter 'a'
A = ord('A')#definimos el codigo UUNICODE para el caracter 'A'
rango = a - A#definimos este rango lo que significara que si se lo restamos a  
#              un caracter en minuscula este me dara el codigo UUNICODE de 
#              dicho caracter en mayuscula
for i in range(len(cadena)):#iteramos en el rango de la cadena de texto
    if cadena[i].isupper():#decimos que si el elemento de la cadena es un
        #                   caracter en mayuscula
        mayuscula1 = cadena[i]#creamos una variable que es igual al elemento de
        #                      la cadena
        print(mayuscula1,end='')
    if cadena[i].islower():#si el elemento de la cadena es un caracter en 
        #                   minuscula
        letra = ord(cadena[i])#creamos una variable que me retore el codigo 
        #                       UUNICODE de dicho caracter
        mayuscula = letra - rango#creamos una variable que al caracter anterior
        #                         le reste el rango encontrado anteriormente
        #                         para de este modo tener el codogo UUNICODE
        #                         de dicho caracter pero en mayuscula
        mayuscula = chr(mayuscula)#cambiamos el codigo UUNICODE 
        #                           caracter UUNICODE
        print(mayuscula,end='')
    if (cadena[i] == ' '):#si en la cadena de texto aparece un espacio
        espacio = ' '#imprimimos el espacio
        print(espacio)
        
```

```python
#cadenas (ejercicio 11)
texto = input('cadena de texto:' )#definimos una entrada para una cadena 
#                                    de texto
palabra = (input('ingrese varias palabras palabra:' ))
len(texto)
Mletra = ''# creamos cadenas vacias que me definiran las tildes, las mayusculas
Ntildes = ''#los digitos y los espacios vacios
Ndigits = ''
Nvacios = ''
for i in range(len(texto)):#iteramos en el rango de la cadena de texto
    if texto[i].isupper() and (texto[i] in 'AEIOU'):#si el elemento es un 
        #                                     es un caracter en mayuscula
        #                                     y aparte es una vocal
        Mletra1 = texto[i]#creamos la variable correspondiente a dicho elemento
        Mletra += Mletra1#y los vamos sumando

    
    if texto[i].isalpha() and ('áéíóúÁÉÍÓÚ' in texto[i]):#si el elemento es un caracter
        #                                   en minuscula o mayuscula y tiene
        #                                   una tilde
        Ntildes1 = texto[i]#creamos variables correspndientes a este elemento
        Ntildes += Ntildes1#y las sumamos

   
    if texto[i].isdigit():#si ele elemento es un digito
        Ndigits1 = texto[i]#creamos una variable y la sumamos
        Ndigits += Ndigits1

    
    if texto[i] == ' ':#si el elemento es un espacio 
        Nvacios1 = texto[i]#creamos una variable y la sumamos
        Nvacios += Nvacios1

   
    if palabra in texto:#si las palabras que entramos estan en la cadena de 
        #                texto
        print('palabra reservada: ',palabra)
        break                                 #va a imprimir esa palabra
 # a continuacion se imprimira la cantidad de mayusculas, tildes, digitos 
#  y vacios en el texto       
print('numero de vocales mayusculas en el texto: ',len(Mletra))
print('numero de letras con tilde en el texto: ' ,len(Ntildes))   
print('numero de digitos en el texto: ',len(Ndigits))
print('numero de vacios en el texto: ',len(Nvacios))
```
